```python
# Classificando dados numéricos por níveis: Baixo, Médio e Alto
def feature_level(feature):
    plt.hist(df[feature])

    # Atribuir eixos x e y + plotar o título
    plt.xlabel(f"{feature.title()}")
    plt.ylabel("Contagem")
    plt.title(f"{feature.title()} Nível")
    plt.show()

    bins = np.linspace(min(df[feature]), max(df[feature]), 4)
    
    printmd("**Classificação**")
    
    printmd(f"Baixo ({bins[0] : .2f} - {bins[1]: .2f})")
    printmd(f"Médio ({bins[1]: .2f} - {bins[2]: .2f})")
    printmd(f"Alto ({bins[2]: .2f} - {bins[3]: .2f})")

    group_names = ['Baixo', 'Médio', 'Alto']

    df.insert(df.shape[1]-1,f'{feature}-nível', pd.cut(df[feature], bins, labels=group_names, include_lowest=True))
    display(df[[feature, f'{feature}-nível']].head(10))


    # Contar valores
    printmd("<br>**Distribuição**<br>")
    display(df[f'{feature}-nível'].value_counts())


    # Plotar a distribuição
    plt.bar(group_names, df[f'{feature}-nível'].value_counts())
    

    # Atribuir eixos x e y + plotar o título
    plt.xlabel(f"{feature.title()}")
    plt.ylabel("Contagem")
    plt.title(f"{feature.title()} Nível")
    plt.show()
```

```python
#### Teste de normalidade estatística: D’Agostino’s K^2, Shapiro-Wilk, Anderson-Darling
### Vamos utilizar neste dataset as 3 Features do tipo numérica: monthlyCharges, customerCharges e totalCharges
##

# 1. monthlyCharges (D’Agostino’s K^2)
stat, p = stats.normaltest(df['monthlyCharges'])

print('Estatística = %.5f, p = %.3f' % (stat, p))

# Interpretar
alpha = 0.05
if p > alpha:
    print('A amostra parece Gaussiana - falha ao rejeitar H0')
else:
    print('A amostra parece Gaussiana - rejeitar H0')
```

```python
# 2. customerStability (D’Agostino’s K^2)
stat, p = stats.normaltest(df['customerStability'])

print('Estatística = %.5f, p = %.3f' % (stat, p))

# Interpretar
alpha = 0.05
if p > alpha:
    print('A amostra parece Gaussiana - falha ao rejeitar H0')
else:
    print('A amostra parece Gaussiana - rejeitar H0')
```

```python
# 3. totalCharges (Anderson-Darling)
result = stats.anderson(df['totalCharges'])

print('Estatística = %.3f' % result.statistic)

p = 0

for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < result.critical_values[i]:
        print(f'Nível de significância {sl:.2f} % : valor crítico {cv:.3f}, os dados parecem normais  - falha ao rejeitar H0')
    else:
        print(f'Nível de significância {sl:.2f} % : valor crítico {cv:.3f}, os dados não parecem normais - rejeitar H0')
```

```python
# Features numéricas

"""#1.Função para plotar histograma e densidade."""
def plot_histogram(feature):
    
    # Criar objeto
    fig = make_subplots(
        rows=2,
        cols=1,
        print_grid=False,
    subplot_titles=(f"Distribuição de {feature.name} com Histograma", f"Distribuição de {feature.name} com gráfico Densidade"))
    
    # Counter histograma
    fig.add_trace(
        go.Histogram(
            x = feature,
            hoverinfo="x+y"
        ),
    row=1,col=1)

    # Densidade
    fig.add_trace(
        go.Histogram(
            x = feature,
            hoverinfo="x+y",
            histnorm = "density"
        ),
    row=2,col=1)
    
    # Atualizar layout
    fig.layout.update(
        height=800, 
        width=870,
        hovermode="closest"
        )
    
    # Atualizar eixos
    fig.layout.yaxis1.update(title="<b>Concentração dos dados</b>")
    fig.layout.yaxis2.update(title="<b>Densidade(%)</b>")
    fig.layout.xaxis2.update(title=f"<b>{feature.name}</b>")
    return fig.show()
```

```python
# Plotando o gráfico de estabilidade
plot_histogram(df['customerStability'])

printmd("**Distribuição do tipo U-shaped**")
```

```python
# Plotando o gráfico de Cobrança Mensal (monthlyCharges)
plot_histogram(df['monthlyCharges'])

printmd("**A distribuição esta fortemente inclinada para direita**")
```

```python
# Plotando o gráfico de Cobrança total (totalCharges)
plot_histogram(df['totalCharges'])

printmd("**Distribuição em forma de J-shaped**")
```

```python
# Preparação dos Dados para Trainamento e Teste do Dataset

strat_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
for train_index, test_index in strat_split.split(df, df["churn"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]

print('Proporção de rótulos de destino no conjunto de dados original\n')
print(df["churn"].value_counts(normalize=True).sort_index())

print('\nTaxa de rótulos de destino no conjunto de dados de teste\n')
print(strat_test_set["churn"].value_counts(normalize=True).sort_index())
```

```python
# Dataset de treinamento
X = strat_train_set.drop("churn", axis=1)
y = strat_train_set["churn"].copy()

# Dataset de teste
y_test = strat_test_set['churn'].values
X_test = strat_test_set.drop('churn',axis=1)

X.shape, y.shape, X_test.shape, y_test.shape
```

```python
# Cardinalidade das variáveis categóricas: 

# reiniciar cat_cols porque 'accountHolderID' ainda está incluído na variável cat_cols
cat_cols = list(set(X.columns) - set(X._get_numeric_data().columns))
num_cols = list(set(X._get_numeric_data().columns) - set({'bankAccount'}))

# Número de entradas exclusivas em cada coluna com dados categóricos
object_nunique = list(map(lambda col: X[col].nunique(), cat_cols))
d = dict(zip(cat_cols, object_nunique))

print("Número de entradas exclusivas por coluna, em ordem crescente:\n")

sorted_cols_by_cardinality = sorted(d.items(), key=lambda x: x[1])
display(pd.DataFrame(sorted_cols_by_cardinality, columns=['Columns', 'Cardinality']))

print("Total de colunas categóricas",len(cat_cols))
print("Total de colunas numéricas",len(num_cols))

printmd("**<br>O conjunto de dados tem valor máximo de cardinalidade ***4***, que é comparativamente baixo<br>**")
```

```python
# Correlação Heatmap: Pearson’s R ou Kendall’s Tau

# Correlação Matrix
# Somente saída de dados numéricos
corr_matrix = pd.concat([df[num_cols],strat_train_set[["churn"]]],axis=1).corr() # spearman, pearson, kendall

# Configurando a máscara para ocultar o triângulo superior
mask = np.zeros_like(corr_matrix, dtype=np.bool)
mask[np.triu_indices_from(mask)]= True

with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr_matrix,
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap = 'coolwarm',
                      cbar_kws = {'shrink': .4,
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1,
                      vmax = 1,
                      annot = True,
                      annot_kws = {'size': 12})

# Adicionando os nomes das colunas como rótulos
ax.set_yticklabels(corr_matrix.columns, rotation = 0, fontsize=13)
ax.set_xticklabels(corr_matrix.columns, fontsize=13)

sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
plt.show()

printmd("**A estabilidade de clientes (customerStability) está moderadamente correlacionada com a Feature numérica target**")
```
